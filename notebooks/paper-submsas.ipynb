# Extract subMSAs of different lengths
Scalability experiment

```python
from collections import namedtuple
from itertools import cycle
from pathlib import Path
DIRSAVE = Path("/data/analysis-paper/paper-experiment-scalability/submsas")
DIRSAVE.mkdir(exist_ok=True, parents=True)

submsas_per_length = 100
columns = list(range(100,3001,100))# + list(range(1500,3001,500)) 

N_SEQS = 50
N_COL_INTERVALS = len(columns)
N_SUBMSAS = submsas_per_length
NAME = f"{N_SEQS}-sars-cov-2-intervals_{N_COL_INTERVALS}-submsas_each_{N_SUBMSAS}-.txt"

if N_SEQS==20:
    PATH = "/data/analysis-paper/paper-experiment-decomposition/sars-cov-2-20-row_maximal/submsas/20-sars-cov-2-ena_alpha481.txt"
elif N_SEQS==50:
    PATH = "/data/analysis-paper/paper-experiment-decomposition/sars-cov-2-50-row_maximal/submsas/50-sars-cov-2-ena_alpha500.txt" 
else:
    PATH = "/data/analysis-paper/paper-experiment-decomposition/sars-cov-2-100-row_maximal/submsas/100-sars-cov-2-ena_alpha200.txt"
```

```python
SubMSA = namedtuple("SubMSA",["length","start","end"])
larger_submsas = []

with open(PATH,"r") as fp:

    for line in fp.readlines():
        line = line.replace("\n","")
        s,e = line.split("\t")
        s = int(s)
        e = int(e)
        if e-s+1 > 100:
            # print(e-s+1, s, e, sep="\t", )
            larger_submsas.append(SubMSA(e-s+1,s,e))

larger_submsas = sorted(larger_submsas, key=lambda t: t[0])
```

```python
# 20 sequences
import random
random.seed(42)
# random.choice(range(100,200))
```

```python

submsas_to_compute = []

iter_submsas = cycle(larger_submsas)

# for submsa in cycle(larger_submsas):

for length_submsas in columns:
    print(length_submsas)
    count_submsas_length = 0
    submsas_length = []

    while count_submsas_length < submsas_per_length:
        
        submsa = next(iter_submsas)
        if submsa.length > length_submsas:  
            start = random.choice(range(submsa.start, submsa.end - length_submsas))
            end = start + length_submsas - 1 
            # print(start, end, end-start+1, sep="\t")
            length = end-start + 1
            assert length == length_submsas, "Inconrrect length of subMSA"
            submsas_length.append(SubMSA(length, start, end))

        count_submsas_length = len(set(submsas_length))
        print("count subMSAs", count_submsas_length)

    submsas_to_compute.extend(submsas_length)

    print("Number of subMSAs", len(submsas_to_compute))
    print("Counter", count_submsas_length)   

submsas_to_compute = sorted(submsas_to_compute, key=lambda t: (t[0],t[1]))

with open(DIRSAVE.joinpath(NAME), "w") as fp:
    for submsa in submsas_to_compute:
        fp.write(f"""- - {submsa.start}\n  - {submsa.end}\n""")
        print(line)
```
